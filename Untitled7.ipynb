```python
import pandas as pd
import numpy as np
```

```python
import os
print(os.listdir())
```

```python
data=pd.read_csv(
    'train_data.txt',
    sep=":::",
    names=['Title','Genre','Description'],
    engine='python',
    encoding='utf-8'
)
print(data)
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Genre']=df["Genre"].str.lower()
```

```python
df['Genre']
```

```python
import re
def clean_text(text):
  text=str.lower(text)
  text=re.sub(r'[^a-z\s]',"",text)
  return text
```

```python
df['Description']=df['Description'].apply(clean_text)
```

```python
df['Description']
```

```python
x=df['Description']
y=df['Genre']
```

```python
print(x.head())
print(y.head())
```

```python
print(x.shape)
print(y.shape)
```

APPLY TF IDF FOR FEATURE

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(
    max_features=20000,
    ngram_range=(1,3),
    min_df=2,
    max_df=0.9,
    sublinear_tf=True)
```

```python
xtfidf=tfidf.fit_transform(x)
```

```python
print(xtfidf.shape)
```

```python
print(tfidf.get_feature_names_out()[20:])
```

CONVET GENRE TO NUMBER

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
y_encode=encoder.fit_transform(y)
```

```python
print(y_encode[:10])
```

Model


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(xtfidf,y_encode,test_size=0.2,random_state=42,stratify=y_encode)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000,C=2)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```

```python
sample = ["A group of zombies attack a small town at night"]

sample_tfidf = tfidf.transform(sample)

prediction = model.predict(sample_tfidf)

genre = encoder.inverse_transform(prediction)

print("Predicted Genre:", genre[0])
```

```python
samples = [
    "haunted house with ghosts and demons",
    "zombie outbreak kills people in a city",
    "vampire attacks a young woman at night"
]

for s in samples:
    s_tfidf = tfidf.transform([s])
    pred = model.predict(s_tfidf)
    print(encoder.inverse_transform(pred)[0])
```
